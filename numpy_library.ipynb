```
n1 = np.array([10,20,30,40,50])
n2 = np.array([40,50,60,70,80])
```

```
#Arrange this array in a veritcal order.
np.vstack((n1,n2))
```

```
#Arrange this array in a horizantal order.
np.hstack((n1,n2))
```

```
#Arrange this array in a column
np.column_stack((n1,n2))
```

```
#Find the common element of an array
np.intersect1d(n1,n2)
```

```
#Find the differnt elements of n1 array.
np.setdiff1d(n1,n2)
```

```
#Find the different elemenst of n2 array.
np.setdiff1d(n2,n1)
```
